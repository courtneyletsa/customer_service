```python
#import libraries
import pandas as pd
```

```python
#load dataset and inspect it
df = pd.read_excel(r"C:\Users\Administrator\Downloads\cleaning_dataset\Customer Call List.xlsx")
df
```

```python
# delete columns that no needed for analysis
df = df.drop(columns='Not_Useful_Column')
df
```

```python
#drop duplicates
df = df.drop_duplicates()
df
```

```python
#Now let's strip the last name column off unwanted characters
df['Last_Name'] = df['Last_Name'].str.strip('123.../_')
df['Last_Name']
```

```python
#Now we might have to standardize the phone number column to suit the format 000-000-000

#first we need to remove all characters from the dataset
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^a-zA-Z0-9]', '', regex=True)


#After removing all characters, we use lambda to introduce hypen between the numbers
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
df['Phone_Number']
```

```python
# replace all nan-- with nothing
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number']
```

```python
# replace all Na-- with nothing
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df['Phone_Number']
```

```python
#Looks good. Now let's look at the Address column. We have address with street name, State and zip code
# How about we split these into three separate columns

df[['Street Address', 'State', 'Zip_code']]=df['Address'].str.split(',',n=2,expand=True)
df
```

```python
# I am going to drop the Address column now since I will not be needing it

df = df.drop(columns='Address')
df
```

```python
# replace all Yes' responses in column Paying Customer with Y and all No's with N

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer']
```

```python
# replace all Yes' responses in column Do_Not_Contact with Y and all No's with N

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Do_Not_Contact']
```

```python
df
```

```python
#drop all None, NaN, values
df = df.fillna('')
df
```

```python
#Because the call center needs to focus on individuals who want to be cantacted, we need to drop all individuals who did not agree to a call back

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
df
```

```python
#We do something similar for empty phone numbers because we assume they do not want to be contacted as well

for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
df
```

```python
#Sweet we are almost done! Now let us look at the index, it needs fixing. Let's reset it
df = df.reset_index(drop = True)
df
```
